```python
dataset['Rating'] = dataset['Rating'].astype(int)
#change rating from float to int
#print the dataset to check the result
print(dataset.Rating)
```

```python
dataset['Span'] = dataset['Span'].str.extract(r'\((\d{4})', expand=False)

# Optionally: Convert the 'year' column to integers if needed
dataset['Span'] = dataset['Span'].astype(int)


# Optionally: Print the updated dataset to verify the changes
print(dataset.Span)
```

```python
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, 3].values
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```
